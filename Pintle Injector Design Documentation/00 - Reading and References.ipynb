## Chapter 5.3.5 Determine Pressure Levels for the Engine & Feed System

> *Pressure drop in the Injector* 

>As a rule of thumb, this pressure drop ($\Delta p_{inj}$) should be about 20% or more of the chamber pressure ($p_c$) for unthrottled engines.

> This drop isolates chamber-pressure oscilations from the feed system, reducing pressure coupling between the combustion chamber and the feed system which could lead to instabilities or oscillations in the flow that are driven by various in combustion. 

> However, for some pintle-type injectors, we can go as low as 5%. 

> The major source of pressure drop through the injectors is from an increase in dynamic pressure. Small injector orifices accelerate the flow so the high-speed propellant streams can atomize and vaporize more easily. However, some flow losses are associated with forcing the propellant through small holes.




## Chapter 5.4 System Sizing, Design, and Trade-offs

>Injector - distributes the prescribed propellant mass flows to the combustion chamber. We must make sure the propellants properly atomize, vaporize, and mix. A large drop in the flow's static pressure (see Sec. 5.3.5) must occur through the injector orifice to ensure that pressure oscillations in the combustion chamber do not travel into the propellant-feed system.


>*Designing the Propellant Injector*

- **Notes**
- Better injector design gives better performance.

> Analysis of injector designs is complicated because many physical phenomena influence propellant injection, mixing, and combustion, all of which occur simultaneously. When the oxidizer and fuel enter the engine, the flow streams strike one another, shatter, atomize, vaporize, diffusively mix, exchange heat, and undergo violent combustion reactions.

> By analyzing each of the processes as a series of uncoupled actions, we make analysis workable... Thus, we can analyze atomizing and vaporization separately, making the total task of designing an injector much easier to complete.


>A reasonable assumption for large-diameter manifolds:
$$\Delta p = K \frac{1}{2} \rho v^2 $$


Continuity equation for a single injector: 

>$$\dot{m} = \rho vA_{inj}$$

Combining:

>$$A_{inj} = \dot{m} \sqrt{\frac{K}{2\rho \Delta p}}$$

where

$v =$ flow velocity through injector (m/s)

$\dot{m} = $ total mass flow rate (kg/s)

$A_{inj} = $ injector inlet cross-sectional area ($m^2$)

$K = $ head loss coefficient (1.2 for radiused inlet, 1.7 for nonradiused) [not exactly applicable to pintle injector]

$\rho =  $ density of the propellant (kg/s)
$\Delta p = $ desired pressure drop across the injector (Pa)

> Notice we have added a factor K. This *head loss coefficient* takes nonreversible pressure losses into account... These "typical numbers" are for standard injector configurations with typical lengths and cross sections. For specific designs, it is a good idea to perform tests (usually with water) early in a program to accurately determine values for K.

> For an injector to work well - that is, produce small droplets which atomize, vaporize and mix rapidly - the tube diameter or channel width must be small. The velocity must be high so when the streams hit, the fluid inertia force is much greater than the surface tension forces in the fluid. Small drops result.

Blockage Factor:

>$$BF = \frac{Nd_o}{\pi d_t} $$

>where $d_t$ is the diameter of the pintle tip, $d_o$ is the diameter of each orifice, and $N$ is the number of orifices... values typically vary between 0.3 and 0.7 (Escher, 1996)


>A second parameter used to characterize pintle injectors is the total
momentum ratio, TMR, defined as the ratio of the momentum rate of the radial propellant to the momentum rate
of the axial propellant. This is given by

>$$TMR = \frac{(\dot{m}U)_r}{(\dot{m}U)_z}$$

>where $U$ is the propellant velocity and the subscripts $r$ and $z$ denote radial and axial propellant flow, respectively.
Escher cites TRW’s experience to state that a TMR $\approx $ 1 yields the highest engine performance. Data or reasoning
for this has not been released.

## Spray Angle Theory

>Several non-dimensional parameters govern fluid breakup and atomization such as that produced by a pintle
injector. The most important of these is usually taken as the Weber number, We, which is defined as the ratio of
fluid inertia to surface tension. This is expressed by

>$$We = \frac{\rho U^2h}{\sigma} $$

>where $\rho$ is fluid density, $U$ is fluid velocity, $h$ is a characteristic length, and $\sigma$ is the fluid’s surface tension with the
surrounding medium (Radovich, 2010). In the case of a pintle injector, the characteristic lengths are taken as the
radial jet diameter and the thickness of the coaxial sheet. Huang’s (1970) study of the breakup of expanding sheets
formed by the impingement of two jets found that breakup characteristics most strongly depend on $We$.

## Some notes

Momentum ratio (Mass flow rate * velocity of inner flow divided by mass flow rate * velocity of outer flow) should be near unity to produce best results. Based upon info given by TRW.

Blockage Factor typically between 0.3 and 0.7. Can be as hight as 1 if injector has a single slot all the way around for central propellant.

>Experience at PERC is that an injection velocity for LOX of less than 10 m/s (32.8 ft/s) leads to a drop-off in performance. It is postulated that at such low velocities there is insufficient kinetic energy in the injected flow streams to provide for complete atomization and mixing.

*Note: Need to be looking at velocities and verify OF ratio...*

>Typical pintle chamber designs show chamber diameters of 3 to 5 times the pintle diameter.



>$L^*$ is defined as the volume of the combustion chamber divided by the area of the nozzle throat

$L^*$ of 30 recommended by TRW. 

Skip distance is the distance from into the chamber that the outer propellant flows before being impinged upon by the inner propellant.

Skip Ratio is the ratio between the skip distance and the pintle diameter. A ratio of 1 is recommended by TRW, while a skip ratio of 2 is considered 'aggressive'.

Primary concern with too long of skip distance is that it exposes the tip to higher temperatures inside of the combustion chamber.

Some evidence suggests that a higher skip ratio is feasible if the diameter of the chamber is much larger than the diameter of the pintle (more than 5x)
