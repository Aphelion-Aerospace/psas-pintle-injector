# Introduction

The propellant injector is one of the critical components of a rocket engine. If the injector incorrectly designed, the best you can hope for is poor performance. At worst, you may be building another test stand from the ground up. 

There are a variety of different injector types. Each of which perform the same core functions of metering the propellants into the engine at the correct flow rate and mixing them. 

The Portland State Aerospace Society (PSAS) has elected to pursue the development of a pintle type injector for its initial fray into liquid propellant rocketry. This notebook is intended to serve as a crash course in the theory, design, manufacture, testing and lessons learned during the 2018-2019 school year.


## PSAS's first liquid bi-propellant rocket engine attempt

The Portland State Aerospace Society (PSAS) is seeking to launch a liquid bi-propellant rocket to a height of 100 km. One of the principal challenges in this endeavor is the development and testing of the rocket engine and associated subsystems. 


PSAS chose to develop a smaller scale, 2.2kN test engine prior to attempting a 100 km capable design. This is in keeping with PSAS’s technology development philosophy of iterative design where initial designs are built, tested, broken, and iterated towards fully functional and flight ready designs. 


The 2.2kN engine was designed to employ a liquid oxygen (LOX) centered pintle. In this design, LOX flows through the center of the pintle and is sprayed outward in a radial pattern. The fuel enters a chamber surrounding the base of the pintle and flows into the engine through the annular gap between the outer diameter of the pintle and the engine. (See Figures 1 & 2) Inside of the combustion chamber, the two propellants impinge upon each other at right angles. This, ideally, results in a cone shaped spray of well mixed fuel and oxidizer.

>Figure 1: Cross section of 2.2 kN engine. Fuel flow path shown in red, LOX flow path shown in green.


<img src="./Images/pintle_dimensions(1).PNG" align="center" width="500"/>

The choice to use this style of injector was finalized by the 2015-2016 capstone team who designed the 2.2kN engine. The use of a pintle type injector is not specifically required for this size of engine. That stated, the initial 2.2kN engine was designed specifically to work with a pintle type injector, which locked a number of the design parameters which may have otherwise been available. The most noteworthy restrictions imposed are that the injector can only be LOX centered, and has a maximum envelope for the outer diameter of the annular gap.

Some of the design considerations which led to choosing a pintle were:

**Combustion Stability:** According to multiple references [cite][cite] pintle type injectors have a higher combustion stability than other types of injectors. This reduces the risk that the injector will create unstable combustion which can lead to an engine failure.

**Throttling:** Pintle type injectors can be designed such that they can be throttled. Throttling offers some potential future advantages for a 100 km engine. Going with a pintle design from the start, in theory, should make the leap throttling easier. 

**Ease of Design** It was believed that a pintle injector would be relatively easy to design and manufacture when compared with other injector types.



# Theory


## Operational Theory

The pintle injector serves two primary purposes in the engine. First, it delivers the fuel and the oxidizer to the combustion chamber at the correct mass flow rate and pressure. Secondly it mixes the propellants and promotes proper combustion.

In practice, the mass flow rate through the injector is controlled by the input pressure for the fuel and oxidizer upstream of the injector. Ideally, the injector will have a pressure drop of between 15 and 25 % of the engine's chamber pressure. 

>"This [pressure] drop isolates chamber-pressure oscilations from the feed system, reducing pressure coupling between the combustion chamber and the feed system which could lead to instabilities or oscillations in the flow that are driven by variations in combustion." [1]

For pintle injectors, this pressure drop can potentially go as low as 5% [1]. Until PSAS has built some heritage with our pintle designs, it is recommended that the 15-25% range be aimed for. 

According to some of the available research on pintle injectors, the optimum propellant spray angle is $45^\circ$.[2] In theory, this spray angle is achieved when the momentums of the two impinging fluids are equal. See Figure 3.

## Key design Equations 

The governing equations employed for the first approximations in this design were:


### Mass flow rate

The mass flow rate for each propellant is one of the critical design parameters for any rocket engine. Thus it is a critical parameter for the injector.

The mass flow rate is defined as:

\begin{equation}
 \dot{m} = \rho VA
\label{eq:massflow}
\tag{1}
\end{equation} 

Where;

$\dot{m}$ is the Mass Flow Rate, $\rho$ is the density of working fluid, $V$ is the mean fluid velocity, and $A$ is the flow area.

### Energy Equation

\begin{equation}
 p_{1}+\rho \frac{V^2_1}{2}+\rho g h_{1} = p_{2}+\rho \frac{V^2_2}{2}+\rho g h_{2}+ h_{l,major}+h_{l,minor}
\label{eq:energy}
\tag{2}
\end{equation} 

Where;

$p$ is the fluid pressure, $h$ is the elevation of the fluid, $h_{l,major}$ are the permanent pressure losses due to viscous forces and $h_{L,minor}$ are the permanent pressure losses associated with changes to flow geometry, (pipe bends, valves, etc)


### Neglected terms and their assumptions

#### Elevation Pressure

The change in elevation between one side of the injector and the other is small enough as to have no measurable effect with the instrumentation being used. Even if this were not the case, the elevation difference is small enough as to have make no practical difference to the calculation results. 

#### Dynamic Pressure

The dynamic pressure required to increase the fluid velocity from rest (in the tank) to the entrance velocity into the injector is relatively small compared to the other pressures involved. For the 2.2kN design it was on the order of 10-20 kPa (2-3 psi). This term is being neglected for the first pass approximation.

#### Major losses

The losses due to viscous forces within the injector are assumed to be negligible due to the short flow lengths inside of the injector.  


### Pressure loss equation

After removing all of the neglected terms from the energy equation we are left with:


\begin{equation}
 p_{1}= p_{2}+h_{l,minor}
\label{eq:pressure1}
\tag{3}
\end{equation} 

The minor losses in this from can be expressed by the equation;

\begin{equation}
 h_{l,minor}= K_{L}\frac{\rho V^2}{2}
\label{eq:minorloss}
\tag{4}
\end{equation} 

where $K_{L}$ is the pressure loss coefficient.

The pressure loss coefficient is an empirically determined value for predicting the permanent pressure loss in a fluid system
Once the pressure loss coefficient known, it is possible to determine the mass flow rate of each propellant through the injector by measuring the pressure directly upstream of the injector and in the chamber. 

Combining equation 3 and 4 gives the basic form of the pressure loss equation used for this design.

\begin{equation}
P_1-P_2 = K_{L}\frac{\rho V_2^2}{2}
\label{eq:pressure}
\tag{5}
\end{equation}

Re-arranged to solve for the pressure loss coefficient;

\begin{equation}
K_L = \frac{2\Delta P}{\rho V^2}
\label{eq:Kl alt}
\tag{6}
\end{equation}

where $\Delta P$ is the gauge pressure upstream of the injector - atmospheric pressure


By substituting $V = \frac{\dot{m}}{\rho  A}$ from equation 1; 


\begin{equation}
K_L = \frac{2\Delta P}{\rho \big(\frac{\dot{m}}{\rho A}\big)^2}
\label{eq:Kl}
\tag{7}
\end{equation}

where $A$ is the exit area of the injector flow path and $\dot{m}$ is the mass flow rate of the associated fluid.

Equation 7 shows that in order to empirically determine $K_L$, one must know  the exit area(s) of the injector, the mass flow rate and density of the fluid, and the corresponding pressure drop.

The exit orifices can be precisely measured using gauge pins, and the mass flow rate and pressure drop can be measured in controlled experiments. 


Equations 6 and 7 were used for the loss coefficients within the jupyter notebooks developed on this project. This was largely a matter of preference. 

In some references related to injectors, a discharge coefficient $C_d$ is used instead of the pressure loss coefficient. 

The typical relationship between the loss and discharge coefficient is;


\begin{equation}
C_d = \frac{1}{\sqrt{K_L}}
\label{eq:CdKl}
\tag{8}
\end{equation}

When using the discharge coefficient, the equation for calculating the mass flow rate is;


\begin{equation}
\dot{m} = AC_d \sqrt{2\rho(\Delta P)}
\label{eq:massflow cd}
\tag{9}
\end{equation}

which can be rewritten as;

\begin{equation}
C_d = \frac{\dot{m}}{A \sqrt{2\rho(\Delta P)}}
\label{eq:Cd}
\tag{4}
\end{equation}




### Momentum Ratio

According to available research, the spray angle of the mixed propellants has an effect on engine performance.[2] Optimum combustion is supposed to occur when the spray angle is approximately $45^\circ$. Theoretically this happens when the ratio of the momentum rates of the two propellants is equal to 1. 

This ratio can be expressed as;


\begin{equation}
\frac{\dot{m_f}V_f}{\dot{m_o}V_o} = 1
\label{eq:momentum}
\tag{7}
\end{equation}

By substituting the velocity in the momentum ratio for the mass flow rate as defined in equation \ref{eq:massflow}, the optimum momentum ratio equation can be rewritten as;


\begin{equation}
\frac{\dot{m_f}^2}{\rho_f A_f} = \frac{\dot{m_o}^2}{\rho_o A_o}
\label{eq:momentum alt}
\tag{8}
\end{equation}


The calculations for the momentum ratio are slightly non-intuitive because the area used in the calculation is not always the entire exit area. For this style of pintle, the areas used for the momentum ratio is the exit area of the radial pintle holes and the collision area as shown in Figure 4.

For the V1, V2, and V3 pintle designs, the exit areas used to calculate for the momentum ratios were:

Pintle: $\pi \frac{d_p^2}{4}$

Annulus: $d_p\times d_g$

Where; $d_p$ is equal to the diameter of the exit orifices of the pintle, and $d_g$ is equal to the annular gap defined as:

$d_g = \frac{1}{2}(A_{OD} - P_{OD})  $

Where $A_{OD}$ is the outer diameter of the Annulus, and $P_{OD}$ is the outer diameter of the pintle.

<img src="./Images/momentum_ratio_diagram.PNG" align="center" width="500"/>

> Figure 4: Horizontal cross section view of pintle tip. Oxidizer flow path travels radially out from the center and is shown in green. Fuel flow path flows perpendicular to the cross section cut and is indicated by the red and blue striped ring between the Annulus OD and the Pintle OD. The areas of red show where the fuel does not impinge on the oxidizer flow path. Areas in blue show were the fuel and oxidizer collide. An approximation of the collision area is used for the purposes of calculating the momentum ratio.


### Weber Number

The Weber number is the ratio between inertial forces and surface tension.

\begin{equation}
\frac{\rho V^2 D}{\sigma}
\label{eq:Weber}
\tag{9}
\end{equation}

Where; $L$ is a characteristic length (exit diameter of holes, or annular gap) $\sigma$ is the surface tension

The weber number is an important parameter in the formation of droplets. When a fluid jet has a high weber number at its exit, the inertial forces cause the fluid to break up into small droplets. This is the basic principal behind spray bottle atomizers. 

It is suggested that the exit velocity of LOX remain above 10 m/s in order to obtain a sufficiently high Weber number for acceptable droplet formation. [3] The fuel should seek to have a weber number at least as high as that of the oxidizer.


### Reynolds Number
- justifies use of turbulent modeling for CFD

\begin{equation}
\frac{\rho V D}{\mu}
\label{eq:Reynolds}
\tag{10}
\end{equation}

Where; $D$ is a characteristic length (exit diameter of holes, or the hydraulic diameter of the annulus) $\mu$ dynamic viscosity



# Design 

## Reality isn't always convenient

### Loss coefficients

Determining the loss coefficients for the was one of the first challenges to be overcome in designing a pintle type injector. 

The initial strategy employed in this project was to break the internal flow paths of the injector into sections and estimate loss coefficients for each section. 

<img src="./Images/LOX_flow.PNG" align="center" width="500"/>

> Figure 5: Cross section view of V1 pintle with locations of initial loss coefficients used for LOX flow path

<img src="./Images/Fuel_flow.PNG" align="center" width="500"/>

> Figure 6: Cross section view of V1 pintle with locations of initial loss coefficients used for fuel flow path.

The idea was that each flow path within the injector has multiple changes to flow area between point of measurement and the atmosphere. The early hand calculations in this project attempted to approximated the overall loss coefficient by segregating the internal flow geometry into segments which could examined individually for pressure losses. 

The sum of those pressure losses could then be added together to determine a total pressure drop estimate, $\Delta P$. This $\Delta P$ could then be used along with the mass flow rate to estimate a total $K_L$ for the injector. This showed early promise with the V2 injector and gave pressure drops which agreed with CFD models and physical testing. When attempts were made to scale up the injector design to meet the general requirements of an engine capable of reaching 100km, the errors in the estimate quickly diverged from CFD models. 




A note of caution, hand calculations and CFD should be employed during the design phase, but any injector which is to be used in an engine needs be specifically tested to determine its loss coefficients and spray angle performance. Small variations in hole sizes due to manufacturing defects, burs, etc, can effect the actual performance of an injector. Theoretical models are valuable to steer the design into the right neighborhood, but still require physical testing for validation.

As an additional issue, the loss coefficients for this design don't scale linearly with regard to the exit areas. This is especially true for the radial orifices of the pintle. Generally the wall thickness of the pintle is not increased at the same scale as the exit orifices. As the exit area and mass flow rates are scaled for a larger engine, flow separation in the corners begins to reduce the effective exit area. This separation can be seen in Figure 7. 


### Exit Velocity

Flow separation in the tip of the injector can reduce the actual flow area of the orifice by a significant amount. This increases the velocity and messes with the momentum ratio calculations. Because of this, $v = \frac{\dot{m}}{\rho A}$ is not always accurate. (See Fig 7)

# Manufacturing

## Test articles

Any new injector needs to be tested and validated prior use in a engine. Early physical testing can and should be done with water in order to experimentally verify the loss coefficients and spray angle. 

The need for testing and verification is especially important when exploring a new injector design, or a design which has been altered to obtain a different mass flow rate or mixture ratio. 

During the design phase, attention should be given to finding ways to produce dimensionally accurate, lower cost, prototypes which can be made, tested, and iterated upon **quickly**. The faster and cheaper an prototype can be made, the faster the development process will go.

As of the 2018-2019 capstone several different prototyping methods have been attempted in the development of the 2.2kN V series injectors.


### Some critical dimensions

Pintle injectors have certain dimensions which require very high precision in order to function properly. 

#### Concentricity

Ensuring that the pintle is correctly centered in the annulus is particularly important. Even small amounts of concentricity error can lead to uneven exit velocity, a lopsided spray angle, and less than optimum combustion.

#### Exit orifices

If the pintle is to employ circular orifices, the as-built dimensions should be measured using gauge pins. The As-built dimensions should thereafter be used during testing and evaluation of the design. This includes re-running any hand calculations with the exact exit areas measured.

#### Gaps

For the annular gap found in the V series design, some success was had using gauge pins at multiple locations around the pintle to determine both concentricity and the exact gap distance. There may be better, more accurate ways to do this which should be looked into if at all possible.


### Machined Aluminum


The V1 and V2 injector assemblies were both manufactured in aluminum. These required students who had machining experience in order to make and were time consuming to produce. They required very careful machining and assembly. 

<img src="./Images/Pintle_V1_V2_compare.jpg" align="center" width="500"/>

> Figure 8: Machined aluminum V1 and V2 pintle injectors (right and left respectively) 

The particular design style for these injectors has room for improvement when it comes to manufacturability. 

Aluminum is a relatively soft metal, which made it very easy to accidentally strip any internal threads which were tapped into the aluminum. This particular issue could be solve by using helicoil threated inserts for any tapped holes.  

Also, the use of shoulder bolts to center the pintle was tried for both the V1 and V2 pintles. The V3 pintle employed countersunk screw heads which made centering the pintle into the annulus much easier to accomplish during assembly. 


### 3D printed

#### Plastics

Most 3D printed plastics which were available on campus at the time of this writing are not suitable for water testing injectors. 

The closest type of 3D print to being workable was done with on a Formlabs printer. It could possible be done, but the material is very brittle and easily breaks when holes are drilled into it, or if you attempt to tap threads. 


<img src="./Images/v3_broken_prints.PNG" align="center" width="500"/>

> Figure 9: Broken V3 injector print created on a Formlabs printer. The part broke when NPT threads were attempting to be tapped into it. Don't do this. It doesn't end well. Oversize the hole in the print and epoxy in metal fittings.


Using a Formlabs could potentially work if you oversize the entrance holes and epoxy in threaded hardware. You will need to pay close attention to the dimensional accuracy. You should pay special attention to checking the the circularity of the pintle and the concentricity of the annulus. If the pintle is not printed vertically, it can come out slightly lop sided. The same goes for the annulus. This can pose an issue for trapped liquid in this type of resin printer. Careful designing would need to take place in order to get it to work, and all dimensions would need to be measured and verified, preferably with gauge pins and micrometers.


#### MJP from 3D systems

When it comes to plastics, Multi-Jet Printing (MJP) is by far the preferred option. Multiple prints of the V3 injector were made using MJP and they gave excellent results. 

<img src="./Images/Pintle_V3_prints.jpg" align="center" width="500"/>

> Figure 10: 3D prints of the V3 injector by 3D systems using the MJP process. Each of the prints created had very high dimensional accuracy and tight tolerances. 


They are not as strong as metal, but the first set of printed injectors were tested up to 60 psi and performed extremely well. 

The prints have very high dimensional accuracy and can have NPT pipe threads printed into them directly without the need to separately tap them. The same goes for helicoil taps. It is still recommended that helicoil or other threaded inserts be used if parts are to be mated with machine screws. They are still plastic and will strip easily if using metal screws on bare plastic. 

Design, tolorance, and technical information can be found on the 3D systems website.


#### Other options

Some other possible options where were not attempted are 3D printing in metal and / or coming up with an easier to manufacture design which could be done quickly, and in house. For instance, it may be possible to find some commercial off the shelf (COTS) part(s) which could modified to serve as a portion of the injector. This could greatly speed up the manufacturing of test articles. It also could potentially lower the development costs.

As well, finding ways to replace as few parts as possible between iterations should be considered. An example could be to create a swappable pintle tip which can allow the testing of different orifice sizes and patterns, without having to rebuild the entire injector. 


## Final injector

The first major concern for the final version of an injector is material selection. The injector is going to be subjected to liquid oxygen, high temperature combustion products, and relatively high pressure. The tip of the pintle is exposed to extreme heating which can potentially damage the pintle and make it a single use item. Thought should be put into how to either make the pintle tip easily replaced between tests, or how to keep it from getting too hot, preferably both.

The metal for the pintle needs to be LOX compatible: Generally this means 304 or 316 Stainless steel, with 304 having a slightly higher melting temperature. There are other material options as well such as Inconel (1390-1424$^\circ$C), Nickel, Copper, or Nickel-Copper Alloys.  

Aluminum is LOX compatible, but will not take the extreme heat which is expected on the pintle tip. (melting point ~ 660$^\circ$C)

For example it may be possible to 3D print the injector in Inconel. It would have its own manufacturing challenges when it comes to post processing, but it could potentially be done.


# Hydro Testing

## Purpose and Theory 

Once an injector design has been manufactured and checked to ensure all of the components fit properly, the first round of testing is performed with water.


## Deliverables

### Loss Coefficients
The first key deliverable from this test is to determine the loss coefficient for each propellant flow path through the injector.

### Spray Angle 
The second deliverable is verification that the injector will create a 45$^\circ$ spray pattern with acceptable mixing and droplet formation.

### Momentum Ratio
The third deliverable is determining if the injector meets the momentum ratio requirement for the propellant mass flow rates.  


## Testing equipment

- Pressure transmitters
    - Go for 0-100 psi transmitters for hydro testing. Using transmitters with a larger range will only decrease the resolution of the measurements and lower the accuracy.
    
- Bucket and a stop watch
- Load Cell capable of giving mass vs time data would be even better

## Testing Procedures

- The full testing procedure should be its own document, this should be a rough summary. 

## Analysis of the results

### Loss Coefficients

- Convert volume flow rate of water in gpm or liters/min to mass flow rate in kg/s 
- convert pressure from psi to pa
- solve loss coefficient equation (7) using mass flow rate, pressure, density of water (measure the water temp, or estimate based on available data on typical tap water temp), and exit flow areas based on the measured as-built exit geometry.

### Spray Angle 

- measure the angle, adjust the flow until you get a good 45 degree spray, then do full tests to determine what the pressure drops for each flow path which create that angle.

### Momentum Ratio

- slightly more complicated to solve for, but with the pressure drops from the spray angle tests along with the loss coefficients from the loss coefficient tests, should be able to work out what the momentum ratio would be for actual propellants selected for engine design.

# LV4 Injector design ideas and suggestions; going for 100km

## Some rough requirements

- 1.5 kg/s LOX
- 1.1 kg/s IPA (70%)

## The LV4 injector as a scaled up 2.2kN design

The first attempt to explore an injector design within the approximate flow range of which may be needed by LV4 yielded results that were frustrating and initially confusing.

This first attempt sought to scale up the V3 2.2kN injector on the naive assumption that the loss coefficients and exit velocities would scale in a predictable enough fashion to be able to roughly calculate the correct size based upon the physical testing and CFD results for the 2.2kN injectors. This assumption was incorrect.

As stated earlier, both the loss coefficients and the exit velocities did not behave as originally assumed.  This prompted a series of full factorial experiments using CFD, where key dimensions and parameters were systematically adjusted to isolate the key parameters that affect the loss coefficients for this design. The results of the experiments were then used to create a set of piecewise curve fit equations that can be used to obtain a good estimate of the loss coefficients for any size injector between the V1 2.2kN injector with approximately 0.2 kg/s total propellant mass flow rate, up to an injector with over 2.5 kg/s total propellant mass flow rate.

It was while doing this analysis and experimenting that it became apparent that the original pintle design will not work for a 6 or 8 kN engine without some modifications to address the issues created by flow separation in the radial orifices.


<img src="./Images/pintle_x_section_flow_separation_2.PNG" align="center" width="500"/>

> Figure 14: CFD image showing flow velocity in a cross section of with modified pintle tip. Mass flow rate was set to 1.5 kg/s. 

The addition of an inverted cone shape to the inside of the pintle tip reduced the pressure drop in the CFD model from 180 psi down to 100 psi for the same 1.5 kg/s flow rate. However, the exit velocity of the fluid actually increased from 12 m/s to nearly 14 m/s. As can be seen in Figure 7, the modified design is does not seem to fix the flow separation issue. The momentum ratio is also negatively impacted by this increase in exit velocity of the denser fluid, which also has a higher mass flow rate requirement.   

While it may be possible to get this design to work, it creates makes obtaining the optimum momentum ratio challenging at best. 

### Alternative design possibilities to explore

Some potential design changes which could be made:

#### Add more holes
Add a second row of orifices for the pintle. The secondary orifices should be offset from the initial row of orifices. This could increase the blockage factor and increase the overall exit area for the pintle. 


#### Rethink the pintle design

Redesign pintle to employ a radial sheet spray pattern rather than using radial orifices.
