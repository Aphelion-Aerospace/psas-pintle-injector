# Flow Calculations and Analysis for Pintle V1


Flow calculations were performed for the as-built Pintle V1 as part of validating our initial design. 

When searching for the original notes and design documentation for the Pintle V1 it was found that the numbers and calculations used were not clearly documented. We were able to verify that the pintle V1 was machined according to the V1 CAD model. To perform the calculations a clean sheet approach was taken using dimensions from the CAD models and the mass flow requirements from the engine design. 

For this notebook, Pintle refers both to the physical injector, and to the inner flow path(s) which will be taken by the Oxidizer. Annulus refers to both to the gap between the outer diameter of the pintle and the engine, as well as the flow path to be taken by the fuel.

# Calculation Objectives

The primary objective of these calculations is to estimate the pressure required to obtain the required engine flow rates with the current Pintle design. 

Secondary objectives are; 
- building a framework for estimating the design parameters for iterating the current injector design
- providing a sanity check for computer modeling the injector using Computational Fluid Dynamics


# Engine Requirements

The pintle injector needs to meet the following requirements:

1. Oxidizer: Liquid Oxygen (LOX)
2. Fuel: 70% isopropyl alcohol, 30% water (IPA)
3. Meet mass flow rates at a pressure differential between 15-20 % of chamber pressure: 
    - Engine chamber pressure 350 psi (as designed)
    - Target pressure differential: 53-70 psi   
4. LOX mass flow rate of 0.9 lbm/s (0.4082 kg/s)
5. Fuel mass flow rate of 1.16 lbm/s (0.5262 kg/s)


# Governing Equations

Given the high expected Reynolds numbers, it is being assumed that the pressure losses will be dominated by the inertial effects, rather than viscous effects.

The governing equations employed for the first pass approximations:


Mass flow rate:

\begin{equation}
 \dot{m} = \rho VA
\label{eq1}
\tag{1}
\end{equation} 

Where;

$\dot{m} =$  Mass Flow Rate (from flow meter) 

$\rho = $ Density of working fluid 

$V = $ Mean fluid velocity

$A = $ Flow area

Energy Equation, with elevation, velocity, and head loss due to friction neglected:

\begin{equation}
P_1-P_2 = \sum K_L\frac{\rho V_2^2}{2}
\label{eq2}
\tag{2}
\end{equation}

Where;

$K_L = $ Pressure Loss coefficient  
$P_1-P_2 =$ Pressure drop 


## Setting up the problem

The velocity at any point in the injector can be determined by rearranging the mass flow equation such that;

\begin{equation}
V = \frac{\dot{m}}{\rho  A}
\label{eq3}
\tag{3}
\end{equation}

Where the mass flow rate, density, and flow areas are all set by the design requirements.

For the analysis, both the LOX and Fuel flow paths were divided into flow regions based upon the part geometry. Each of the regions have a particular flow area for which the velocity of the flow can be solved.

At every change in geometry a loss coefficient $K_L$ was added to account for associated the dynamic pressure losses.
