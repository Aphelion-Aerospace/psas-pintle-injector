# Pintle Test for 2.2kN Liquid Propellant Rocket Engine
Dates of testing: 19 & 26 Jan 2019

# Introduction


The Portland State Aerospace Society is in the process of testing its $2.2kN$ ($500 lb_f$) regeneratively cooled bi-propellant rocket motor. One of the critical steps prior to hot firing the motor is the testing and validating of the injector using water.

The intitial test design was set up to only require a single turbine flow meter and an orifice plate in order to determine flow rate for each of the two propellant paths. During calibration it was found that the flow meter was not working reliably. It was decided to attempt to test the pintle using only the calibrated orifice plate and back solving for flow rates as described in the methods section.

## Underlying theory behind the experimental set up.

A calibrated orifice with two pressure transducers was used to determining the total mass flow rate of the water through the circuit. The discharge coefficient was determined using the following equation:

\begin{equation}
C_d = \frac{Q_{gpm}}{\sqrt{\Delta P}}
\label{eq1}
\tag{1}
\end{equation}

Where;

$C_d = $ the discharge coefficent 

$Q_{gpm} =$ is the volumentric flow rate in gallons per minute 

$\Delta P = $ is the difference in pressure across the orifice 

The orifice was calibrated using multiple input pressures. The flow rate was determined using a stop watch and a graduated container. Times were recorded every quarter gallon and then plotted to determine flow rates. Each of the flow rates was then plotted against the corresponding pressure drop and compared to the experimentally derived $C_d$ for the main orifice. (figure)

<img src="main_orifice_calibration_graph.png" alt="Drawing" style="width: 700px;"/>


<center>Figure 1: Pressure Loss calibration curve and experimental data for calibrated orifice. Experimental data indicated that the accuracy of the orifice drops off steeply at pressure differences over 30 psi</center>


Once the main orifice is calibrated, the pressure difference can be used to determine the total flow rate for testing each fluid circuit. 

Using the orifice plate pressure drop combined with the pressure drop of pressure transducer 1 (fuel side) a discharge coefficent for the fuel annulus can be determined.

This method can then be repeated to determine the discharge coeffienct for the radial orifices of the LOX circuit.


Each of the two fluid circuits was then tested separately in order to determine the loss coefficients across the pintle for both the LOX and Fuel sides. 
After determining the loss coefficients for the main orifice, both propellant paths across the pintle we could then determine the flow rates during spray angle tests using pressure data.
